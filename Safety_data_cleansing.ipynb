## Import dataset

```python
## import data from feature folder, I renamed the csv name 
data1 = pd.read_csv('/Users/a/Desktop/safety/features/part1.csv')
data2 = pd.read_csv('/Users/a/Desktop/safety/features/part2.csv')
data3 = pd.read_csv('/Users/a/Desktop/safety/features/part3.csv')
data4 = pd.read_csv('/Users/a/Desktop/safety/features/part4.csv')
data5 = pd.read_csv('/Users/a/Desktop/safety/features/part5.csv')
data6 = pd.read_csv('/Users/a/Desktop/safety/features/part6.csv')
data7 = pd.read_csv('/Users/a/Desktop/safety/features/part7.csv')
data8 = pd.read_csv('/Users/a/Desktop/safety/features/part8.csv')
data9 = pd.read_csv('/Users/a/Desktop/safety/features/part9.csv')
data10 = pd.read_csv('/Users/a/Desktop/safety/features/part10.csv')

df = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10],axis=0) ##concatenate all data
df = df.sort_values(by=['bookingID','second'],ascending=True)  ## sort values first by booking_id then by seconds 
```

```python
df.describe()
```

```python
## clean the wrong second record
df=df.loc[df.second < 43200]  ## trip less than 12 hours 
df=df.loc[df.Speed > 0] # speed larger than 0 and less than 200km/h
df=df.loc[df.Speed < 55.56] # speed larger than 0 and less than 200km/h
df=df.loc[df.Accuracy < 50] ## ensure accuracy level
```

```python
df.describe()
```

```python
##groupby bookingid then generate a list for each field
df = df.groupby(['bookingID']).agg(lambda x: list(x)) 
```

```python
## the csv file is renamed as label
label = pd.read_csv('/Users/a/Desktop/safety/labels/label.csv') 
## sort the dataframe so that for duplicate bookingid, the label associated with the last record is always 1(dangerous)
label = label.sort_values(by = ['bookingID','label'],ascending=True) 
## drop the duplicate values except the last occurance, if the bookingid has both label 0 and 1, we treat it as 1(dangerous) because it is suspicious
label = label.drop_duplicates(subset=['bookingID'],keep='last') 
## set bookingID to be the index
label = label.set_index('bookingID',drop = True)
```

```python
##join features and labels
df_new = pd.concat([df,label],axis=1)
```

```python
df_new = df_new.dropna(axis=0)
df_new.info()
```

```python
## export the merged dataframe for future use
import pickle
with open("/Users/a/Desktop/df_raw.pkl","wb") as handle:
    pickle.dump(df_new,handle,protocol=pickle.HIGHEST_PROTOCOL)
```
